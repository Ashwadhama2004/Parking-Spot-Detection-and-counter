```python
import cv2
```

```python
video_path="Parking spot detection and counter\data\parking_crop.mp4"
```

```python
cap = cv2.VideoCapture(video_path)
```

```python
ret = True

while ret:
  ret, frame = cap.read()
  cv2.imshow('frame', frame)
  if cv2.waitKey(25) & 0xFF == ord('q'):
    break
```

```python
cap.release()
```

```python
cv2.destroyAllWindows()
```
